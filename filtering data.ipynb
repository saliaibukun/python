```python
import pandas as pd

data = {
    'name': ['alice','bob','charles','david','eva'],
    'age': [2,37,30,22,55],
    'gender': ['F','M','M','M','F'],
    'salary': [50000,60000,55000,80000,45000]
    }
df = pd.DataFrame(data)
print(df)
```

```python
condition = df['age'] >= 30
print(condition)
```

```python
filtered_df = df[condition]
print(filtered_df)
```

```python
filtered_df = df[(df['age'] >= 37) & (df['gender'] == 'M')]
print(filtered_df)
```
