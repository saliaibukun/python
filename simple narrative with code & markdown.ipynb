```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female'],
    'Race/Ethnicity': ['Group A', 'Group B', 'Group C', 'Group B', 'Group A'],
    'Parental Education': [
        "Bachelor's degree", 
        "Some college", 
        "Master's degree", 
        "High school", 
        "Associate's degree"
    ],
    'Lunch': ['Standard', 'Free/reduced', 'Standard', 'Free/reduced', 'Standard'],
    'Test Prep Course': ['Completed', 'None', 'Completed', 'None', 'Completed'],
    'Math score': [85, 72, 90, 66, 95],
    'Reading score': [88, 79, 85, 70, 92],
    'Writing score': [90, 74, 88, 68, 94]
}
df = pd.DataFrame(data)

df
```

Dataset Description

The dataset contains information about students performance in three subjects:
- Maths
- Reading
- Writing

Each student entry also includes gender, race/ethnicity, parental education, lunch type and test preparation course.

```python
df.describe()
```

## Observation

- Average math score is around 81.
- There is some variation across subjects.
- Let's now visualize the score distribution


```python
sns.histplot(df['Math score'], bins=20, kde=True,color='skyblue')
plt.title("Distribution of math score")
plt.xlabel("score")
plt.ylabel("frequency")
plt.show()
```

```python
avg_scores_by_gender = df.groupby('Gender')[['Math score','Reading score','Writing score']].mean()
avg_scores_by_gender
```

## Gender-Based Average Scores

from the table above, we observe :
- Female tend to score higher in math.
- Female also tend to perform better in both reading and writing.

## Conclusion

- There is a clear pattern in performance based on gender.
- Preparation courses seems to improve scores.
- Visualizing data help to understand hidden trend easily.
