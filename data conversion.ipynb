```python
import pandas as pd 
```

```python
data = {
    'age': ['45','22','30'],
    'salary': [40000,50000,60000],
    'join_date': ['2025-01-01','2025-07-15','2025-04-20']
    }
df = pd.DataFrame(data)
print(df.dtypes)
```

```python
df['age']= df['age'].astype(int)
```

```python
df['join_date']= pd.to_datetime(df['join_date'])
```

```python
print(df.dtypes)
```

```python
df['height'] = ['5.8','6.1','5.5']
df['department'] = ['HR','finance','HR']
df['experience'] = ['2','three','4']

 
print(df)
```

```python
df['height'] = df['height'].astype(float)
```

```python
print(df.dtypes)
```

```python
df['department'] = df['department'].astype('category')
```

```python
print(df.dtypes)
```

```python
df['experience'] = pd.to_numeric(df['experience'],errors='coerce')
```

```python
print(df.dtypes)
```

```python
print(df)
print("\nData types:\n", df.dtypes)
```
