```python
import pandas as pd
import numpy as np

data = {
    'name': ['alice','bob','charles','david','eva','alice'],
    'age': ['25','np.nan','30','29','NaN','25'],
    'gender': ['female','male','male','np.nan','female','female'],
    'country': ['USA','usa','canda','CANADA','UK','USA'],
    'quantity': [2, 1, 3, 5, 2, 2],
    'unit_price': [20,15,np.nan,10,30,20] 
}

df = pd.DataFrame(data)
df
```

```python
df.fillna(0)
```

```python
df.dropna()
```

```python
df.drop_duplicates()
```

```python
df['country'] = df['country'].str.lower()
print(df)
```

```python
df['age'] = pd.to_numeric(df['age'],errors = 'coerce')

df['age'] = df['age'].fillna(df['age'].median())
df['age']= df['age'].astype(int)

print(df)
```

```python
df['total_price'] = df['quantity'] * df['unit_price']

print('total_price')
```
