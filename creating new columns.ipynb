```python
import pandas as pd
import numpy as np
```

```python
data = {
    'name': ['alice','bob','charles','david'],
    'maths': [88,56,92,88],
    'science': [90,82,85,95]
    }
df = pd.DataFrame(data)
print(df)
```

```python
df['average'] = (df['maths'] + df['science']) / 2
print(df)
```

```python
df['results'] = df['average'].apply(lambda x: 'pass' if x > 80 else 'fail')
print(df)
```

```python
df['year'] = 2026
print(df)
```

```python
def grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    else:
        return 'C'


df['grade'] = df['average'].apply(grade)
print(df)
```

```python
df['maths_level'] = df['maths'].apply(lambda x: 'high'if x > 80 else 'low')
print(df)
```

```python
df['science_level'] = np.where(df['science'] >=90,'Top', 'Average')
print(df)
```

```python
df['grade'] = df['maths'].apply(lambda x: 'A' if x>=90 else ('B' if x>=80 else 'C'))

grade_map = {
    'A': 'excellent',
    'B': 'good',
    'C': 'needs improvement'
}

df['grade_description'] = df['grade'].map(grade_map)
print(df)
```
