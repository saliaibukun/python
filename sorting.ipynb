```python
import pandas as pd
```

```python
data = {
    'name': ['alice','bob','charles','david'],
    'age': [25,56,33,24],
    'score': [85,90,100,88],
    'country': ['USA','Canada','UK','UAE']
    }
df = pd.DataFrame(data)
print(df)
```

```python
df_sorted_age = df.sort_values(by='age')
print(df_sorted_age)
```

```python
df_sorted_score = df.sort_values(by='score')
print(df_sorted_score)
```

```python
df_sorted = df.sort_values(by=['score','age'], ascending=[False,True])
print(df_sorted)
```

```python
df_sorted_index = df.sort_index()
print(df_sorted_index)
```

```python
df.set_index('name', inplace=True)
print(df.loc['charles'])
```

```python
print(df.iloc[3])
```

```python
print(df.loc[['charles','bob'],['age','score','country']])
```

```python
df_reset = df.reset_index()
print(df_reset)
```
