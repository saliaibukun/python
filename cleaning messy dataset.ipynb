```python
import pandas as pd
import numpy as np
```

```python
data = {
    'name': ['alice','bob',None,'david','alice'],
    'age': ['25','thirty','35',None,'25'],
    'salary': [50000,60000,None,55000,50000],
    'join_date': ['2025-01-01','not available','2025-05-01','2025-11-01','2025-01-01']
    }
df = pd.DataFrame(data)
df
```

```python
df.info()
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['age'] = df['age'].replace('thirty', np.nan)
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['age'] = df['age'].fillna(df['age'].mean())

df['name'] = df['name'].fillna('unknown')
df['salary'] = df['salary'].ffill()
```

```python
df['join_date'] = pd.to_datetime(df['join_date'], errors='coerce')
```

```python
df =df.drop_duplicates()
```
